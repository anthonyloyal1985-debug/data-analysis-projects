### **Simple Data Analysis**

**Find the total sales from all types of items sold**

```python
total_sales = blinkit_data['Total Amount'].sum()
print("Total sales from all types of items sold:", round(total_sales, 2))
```

```python
print(f"Total sales from all types of items sold: $ {round(total_sales/1000000, 2)} M")
```

```python
print(f"Total sales from all types of items sold: ₹ {round(total_sales * 94.01/1000000 , 2)} M")
```

**Find the average sales from all types of items sold**

```python
average_sales = blinkit_data['Total Amount'].mean()
print("Average sales from all types of items sold:", round(average_sales, 2))
```

```python
print(f"Average sales from all types of items sold: $ {round(average_sales / 1000, 2)} K")
```

**Find the total number of Transactions**

```python
total_transactions = blinkit_data['Item Identifier'].count()
print(f"Total number of Transactions: {total_transactions}")
```

```python
print(f"Total number of Transactions: {round(total_transactions / 1000, 2)} K")
```

**Finding the average ratings of items sold**

```python
average_rating = blinkit_data['Rating'].mean()
print(f"Average ratings of items sold: {round(average_rating, 2)}")
```

### **Exploratory Data Analysis**

**EDA on Sales by Fat Content**

```python
blinkit_data.groupby('Item Fat Content').agg(
    Total_Sales = ('Total Amount', 'sum'),
    Average_Sales = ('Total Amount', 'mean'),
    Total_Transactions = ('Item Identifier', 'count'),
    Average_Rating = ('Rating', 'mean')
).round(2)
```

**EDA on Sales by Item Type**

```python
blinkit_data.groupby('Item Type').agg(
    Total_Sales = ('Total Amount', 'sum'),
    Average_Sales = ('Total Amount', 'mean'),
    Number_Of_Transactions = ('Item Identifier', 'count'),
    Average_Rating = ('Rating', 'mean')
).round(2)
```

**EDA on Sales by Outlet Location**

```python
blinkit_data.groupby('Outlet Location Type').agg(
    Total_Sales = ('Total Amount', 'sum'),
    Average_Sales = ('Total Amount', 'mean'),
    Number_Of_Transactions = ('Item Identifier', 'count'),
    Average_Rating = ('Rating', 'mean')
).round(2)
```

**EDA on Sales by Outlet Size**

```python
blinkit_data.groupby('Outlet Size').agg(
    Total_Sales = ('Total Amount', 'sum'),
    Average_Sales = ('Total Amount', 'mean'),
    Number_Of_Transactions = ('Item Identifier', 'count'),
    Average_Rating = ('Rating', 'mean')
).round(2)
```

**EDA on Sales by Outlet Type**

```python
blinkit_data.groupby('Outlet Type').agg(
    Total_Sales = ('Total Amount', 'sum'),
    Average_Sales = ('Total Amount', 'mean'),
    Number_Of_Transactions = ('Item Identifier', 'count'),
    Average_Rating = ('Rating', 'mean')
).round(2)
```

**EDA on Sales by Fat Content and Item Type**

```python
blinkit_data.groupby(['Item Fat Content', 'Item Type']).agg(
    Total_Sales = ('Total Amount', 'sum'),
    Average_Sales = ('Total Amount', 'mean'),
    Number_Of_Transactions = ('Item Identifier', 'count'),
    Average_Rating = ('Rating', 'mean')
).round(2)
```

**EDA on Sales by Top 5 Performing Item Types**

```python
blinkit_data.groupby('Item Type').agg(
    Total_Sales = ('Total Amount', 'sum'),
    Average_Sales = ('Total Amount', 'mean'),
    Number_Of_Transactions = ('Item Identifier', 'count'),
    Average_Rating = ('Rating', 'mean')
).sort_values(by = 'Total_Sales', ascending = False).round(2).head()
```

**EDA on Sales by Top 5 Outlets by Threir Establishment Year**

```python
blinkit_data.groupby('Outlet Establishment Year').agg(
    Total_Sales = ('Total Amount', 'sum'),
    Average_Sales = ('Total Amount', 'mean'),
    Number_Of_Transactions = ('Item Identifier', 'count'),
    Average_Rating = ('Rating', 'mean')
).sort_values(by = 'Total_Sales', ascending = False).round(2).head()
```
