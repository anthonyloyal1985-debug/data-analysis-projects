# Numpy Project :- Student Result Analysis

##### **Description About Data**
There is a 'students_marks.csv' file which contains data of 100 students.
There are 100 rows where every row representing a student data. 
There are four columns where first column contains roll number and 
next three columns contains marks of maths, physics and chemistry.

**Importing the numpy library**

```python
import numpy as np
```

**Read the data from csv file into numpy array**

```python
students_data = np.loadtxt("students_marks.csv", delimiter = ",")
```

**Find the shape of data**

```python
students_data.shape
```

**Display first 5 rows of data**

```python
print(students_data[:5])
```

**Find the total of marks of each student**

```python
total_marks = students_data[:,1:].sum(axis = 1)
print(total_marks)
```

**Add a new column that will store the total of marks of each student**

```python
print(total_marks.reshape(1, -1))
```

```python
print(total_marks.reshape(-1, 1))
```

```python
print(total_marks)
```

```python
total_marks = total_marks.reshape(-1, 1)
print(total_marks)
```

```python
students_data = np.concatenate((students_data, total_marks), axis = 1)
```

**Display first 5 rows of data with newly added column for total of marks**

```python
print(students_data[:5])
```

**Find the average of marks of each student**

```python
average_marks = students_data[:,1:-1].mean(axis = 1).round(2)
print(average_marks)
```

```python
average_marks = average_marks.reshape(-1, 1)
print(average_marks)
```

**Add a new column that will store the average of marks of each student**

```python
students_data = np.concatenate((students_data, average_marks), axis = 1)
```

**Display first 5 rows of data with newly added column for average of marks**

```python
print(students_data[:5])
```

**Find the average of marks of maths subject**

```python
print("Average marks of maths:", students_data[:, 1].mean().round(2))
```

**Find the average of marks of physics subject**

```python
print("Average marks of physics:", students_data[:, 2].mean().round())
```

**Find the average of marks of chemistry subject**

```python
print("Average marks of chemistry:", students_data[:, 3].mean().round())
```

**Find the highest marks in maths subject**

```python
print("Highest marks of maths:", int(students_data[:,1].max().round()))
```

**Find the highest marks in physics subject**

```python
print("Highest marks of physics:", int(students_data[:,2].max().round()))
```

**Find the highest marks in chemistry subject**

```python
print("Highest marks of chemistry:", int(students_data[:,3].max().round()))
```

**Find the lowest marks in maths subject**

```python
print("Lowest marks of maths:", int(students_data[:,1].min().round()))
```

**Find the lowest marks in physics subject**

```python
print("Lowest marks of physics:", int(students_data[:,2].min().round()))
```

**Find the lowest marks in chemistry subject**

```python
print("Lowest marks of chemistry:", int(students_data[:,3].min().round()))
```

**Count the number of students who have scored more than or equal to 90 marks in maths subject**

```python
print("Number of students who have scored more than or equal to 90 marks in maths subject:", 
students_data[:,1][students_data[:,1] >= 90].size)
```

**Count the number of students who have scored more than or equal to 90 marks in physics subject**

```python
print("Number of students who have scored more than or equal to 90 marks in physics subject:", 
students_data[:,2][students_data[:,2] >= 90].size)
```

**Count the number of students who have scored average marks more than or equal to 60**

```python
print("Number of students who have scored average marks more than or equal to 60:", 
students_data[:,5][students_data[:,5] >= 60].size)
```

**Count the number of students who have scored average marks less than 50**

```python
print("Number of students who have scored average marks less than 50:", 
len(students_data[:,5][students_data[:,5] < 50]))
```

**Count the number of students who have scored average marks between 80 and 90**

```python
print("Number of students who have scored average marks between 80 and 90:",
len(students_data[:,5][(students_data[:,5] >= 80) & (students_data[:,5] <= 90)]))
```

**Count the number of students who have scored marks more than or equal to 90 in all subjects**

```python
students_scored_more_than_or_equal_to_90_in_all_subjects = students_data[
(students_data[:,1] >= 90) & (students_data[:,2] >= 90) & (students_data[:,3] >= 90)]
print(students_scored_more_than_or_equal_to_90_in_all_subjects)
```

```python
print("Number of students who have scored marks more than or equal to 90 in all subjects:", 
len(students_scored_more_than_or_equal_to_90_in_all_subjects))
```

**Count the number of students who have scored marks more than or equal to 90 in any subject**

```python
students_scored_more_than_or_equal_to_90_in_any_subjects = students_data[
(students_data[:,1] >= 90) | (students_data[:,2] >= 90) | (students_data[:,3] >= 90)]
print(students_scored_more_than_or_equal_to_90_in_any_subjects)
```

```python
print("Number of students who have scored marks more than or equal to 90 in any subject:", 
len(students_scored_more_than_or_equal_to_90_in_any_subjects))
```

**Saving the Students Result Analysis Data to the 'students_result.csv' file**

```python
np.savetxt('student_result.csv', students_data, delimiter = ',', fmt = ["%d", "%d", "%d", "%d", "%d", "%.2f"], 
header = "Roll Number,Maths Marks,Physics Marks,Chemistry Marks,Total Marks,Average Marks", comments="")
print("Student Result is saved successfully")
```
