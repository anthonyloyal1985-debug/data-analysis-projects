# **Visualize Data**

**Plot a bar plot to compare top 5 countries with most Movies**

```python
movies_data = netflix_data[netflix_data['type'] == 'Movie']
movies_by_top_5_countries = movies_data.groupby('country')['show_id'].count().sort_values(ascending = False).head(5).reset_index()
movies_by_top_5_countries
```

```python
plt.figure(figsize = (12, 3))
plt.title("Number of Movies Released by Top 5 Countries")
plt.xlabel("Country Name")
plt.ylabel("Number of Movies Released")
sb.barplot(movies_by_top_5_countries, x = 'country', y = 'show_id', width = .5, hue = 'country')
plt.show()
```

**Plot a bar plot to compare top 5 countries with most TV Shows**

```python
tv_shows_data = netflix_data[netflix_data['type'] == 'TV Show']
tv_shows_data_by_top_5_countries = tv_shows_data.groupby('country')['show_id'].count().sort_values(ascending = False).head(5).reset_index()
tv_shows_data_by_top_5_countries
```

```python
plt.figure(figsize = (12, 3))
plt.title("Number of TV Shows Released by Top 5 Countries")
plt.xlabel("Country Name")
plt.ylabel("Number of TV Shows Released")
sb.barplot(tv_shows_data_by_top_5_countries, x = 'country', y = 'show_id', width = .5, hue = 'country')
plt.show()
```
